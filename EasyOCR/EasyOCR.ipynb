```python
import easyocr
```

```python
reader = easyocr.Reader(['en'],gpu = False)  #'en' denotes english language
```

```python
import PIL #PIL is used to fetch the image from local system
from PIL import ImageDraw
im = PIL.Image.open("1.PNG")
im
```

```python
bounds = reader.readtext('1.PNG')
bounds
```

```python
#Making a function to draw the bounding boxes around the extracted text.
def draw_boxes(image, bounds, color='yellow', width=2):
    draw = ImageDraw.Draw(image)
    for bound in bounds:
        p0, p1, p2, p3 = bound[0]
        draw.line([*p0, *p1, *p2, *p3, *p0], fill=color, width=width)
    return image

draw_boxes(im, bounds)
```

```python

len(bounds)
```

```python
for i in bounds:
    print(i[1])
```
