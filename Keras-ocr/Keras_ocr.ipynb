```
import keras_ocr
from matplotlib import pyplot as plt
```

```
from google.colab import drive
drive.mount('/content/drive/')
```

```
#Downloading the pretrained model of Keras-ocr.
pipeline = keras_ocr.pipeline.Pipeline()
```

```
images = [keras_ocr.tools.read(img) for img in ['/content/drive/MyDrive/OCR-Dataset/5.jpeg','/content/drive/MyDrive/OCR-Dataset/7.jpg','/content/drive/MyDrive/OCR-Dataset/1.png']]
import numpy as np
print(np.shape(images))
```

```
prediction_groups = pipeline.recognize(images)
fig,axs = plt.subplots(nrows=len(images), figsize=(20,20))
for ax,image,prediction in zip (axs, images , prediction_groups):
  keras_ocr.tools.drawAnnotations(image=image,predictions=prediction,ax=ax)
```

```
for prediction in prediction_groups:
  print(prediction)
```
